```python
import pandas as pd 
```

```python
data = pd.read_csv(r'C:\Users\paton\Downloads\movie_rating.csv')
data.head()
```

```python
movie_rating = data.pivot_table(index= ['user_ID'], columns=['title'], values='rating')
movie_rating.head()
```

```python
ToyStoryRating = movie_rating['Toy Story (1995)']
ToyStoryRating.head()
```

```python
Similarmovies = movie_rating.corrwith(ToyStoryRating)
Similarmovies_cleaned = Similarmovies.dropna()
```

```python
df_recommendation_movies = pd.DataFrame(Similarmovies_cleaned, columns= ['correlation'])
df_recommendation_movies
```

```python
rating_count = data.groupby('title')['rating'].count()
rating_count
```

```python
df_recommendation_movies =df_recommendation_movies.join(rating_count.rename('num_rating'))
df_recommendation_movies.head()
```

```python
recommendations = df_recommendation_movies[df_recommendation_movies['num_rating'] > 10]
recommendations
```

```python
recommendations = recommendations.sort_values('correlation', ascending=False).head(10)
recommendations
```

```python
movierating_by_userID1 = data[data['user_ID'] == 1]['title']
movierating_by_userID1 = list(movierating_by_userID1)
movierating_by_userID1
```

```python
recommendations = recommendations.index
correct = 0
for title in recommendations:
    if title in movierating_by_userID1:
        correct += 1
precision = correct / 10
print ("Precision@10 :" , precision)
        
    
```
