```python
from langchain_openai import ChatOpenAI
from langchain_core.messages import SystemMessage, HumanMessage
from langchain_core.prompts import SystemMessagePromptTemplate, HumanMessagePromptTemplate, ChatPromptTemplate
```

```python
chat = ChatOpenAI(model="gpt-4",
                  temperature=0,
                  model_kwargs= {'seed':365},)
```

```python
s_message = SystemMessage(content='''You are kindly and helpful to answers ''')
h_message = HumanMessage(content='''I'ce recently adopted a dog. Can you suggest some dog name?''')
```

```python
respone = chat.invoke([s_message,h_message])
```

```python
respone 
```

# Build ChatOpenAI Templates

```python
TEMPLATE_S = "{description}"
TEMPLATE_H = '''I'ce recently adopted a {pet}. Can you suggest some {pet} name?'''

s_message_template = SystemMessagePromptTemplate.from_template(template=TEMPLATE_S)
h_message_template = HumanMessagePromptTemplate.from_template(template=TEMPLATE_H)
```

```python
chat_template = ChatPromptTemplate.from_messages([s_message_template,h_message_template])
```

```python
chat_template
```

```python
chat_value = chat_template.invoke({'description':'You are kindly and helpful to answers','pet': 'dog'})
chat_value
                                   
```

```python
respon = chat.invoke(chat_value)
respon
```
