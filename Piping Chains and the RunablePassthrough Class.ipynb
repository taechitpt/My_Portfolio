# Piping Chains and the RunablePassthrough Class

```python
%load_ext dotenv
%dotenv 
```

```python
from langchain_openai import ChatOpenAI
from langchain_core.prompts import ChatPromptTemplate
from langchain_core.output_parsers import StrOutputParser
from langchain_core.runnables import RunnablePassthrough
```

```python
chat = ChatOpenAI(model="gpt-4",
                  temperature=0,
                  max_completion_tokens=100,
                  model_kwargs= {'seed':365},)
```

```python
chat_template_tools = ChatPromptTemplate.from_messages(
    [
        ("system", ''' you are helpful to answers'''),
        ("human", '''What are the five most important tools a {job title} needs? Answer only by listing the tools.''')]
)
chat_template_strategy = ChatPromptTemplate.from_messages(
    [
        ("system", ''' you are helpful to answers'''),
        ("human", '''Considering the tools provided a strategy for effectively learning and mastering them: {tools}.''')]
)
```

```python
chat_template_tools
```

```python
chat_template_strategy
```

```python
string_parser = StrOutputParser()
```

```python
chain_tools = chat_template_tools|chat|string_parser
chain_strategy = chat_template_strategy|chat|string_parser
```

```python
print (chain_tools.invoke({'job title' : 'Data Scientist'}))
```

```python
print (chain_strategy.invoke({'tools' : '''1. Python
2. R Programming
3. SQL
4. Tableau
5. Hadoop'''}))
```

```python
# Another solution by combined chains together.
```

```python
chain_tools = chat_template_tools|chat|string_parser| {"tools":RunnablePassthrough()}
chain_strategy = chat_template_strategy|chat|string_parser
```

```python
combined_chain = chain_tools|chain_strategy
```

```python
print (combined_chain.invoke({"job title": 'Data Scientist'}))
```
