# Building Multi-Step LLM Pipelines using LangChain
### Using Prompt Templates, Output Parsers, and RunnablePassthrough

## Environment Setup
Load environment variables such as API keys from a `.env` file to ensure secure configuration.

```python
%load_ext dotenv
%dotenv 
```

```python
from langchain_openai import ChatOpenAI
from langchain_core.prompts import ChatPromptTemplate
from langchain_core.output_parsers import StrOutputParser
from langchain_core.runnables import RunnablePassthrough
```

## Initialize LLM Model
Set up the ChatOpenAI model with desired parameters such as temperature and token limits.

```python
chat = ChatOpenAI(model="gpt-4",
                  temperature=0,
                  max_completion_tokens=100,
                  model_kwargs= {'seed':365})
```

## Define Prompt Templates
Create prompt templates for generating required tools and strategies based on a given job title.

```python
chat_template_tools = ChatPromptTemplate.from_messages(
    [
        ("system", ''' you are helpful to answers'''),
        ("human", '''What are the five most important tools a {job title} needs? Answer only by listing the tools.''')]
)
chat_template_strategy = ChatPromptTemplate.from_messages(
    [
        ("system", ''' you are helpful to answers'''),
        ("human", '''Considering the tools provided a strategy for effectively learning and mastering them: {tools}.''')]
)
```

```python
chat_template_tools
```

```python
chat_template_strategy
```

## Output Parsing
Convert model responses into structured string outputs using StrOutputParser.

```python
string_parser = StrOutputParser()
```

## Build LLM Chains
Combine prompt templates, model, and parser into a pipeline using LangChain Expression Language (LCEL).

```python
chain_tools = chat_template_tools|chat|string_parser
chain_strategy = chat_template_strategy|chat|string_parser
```

## Run the Chains
Execute the pipeline to generate tools and strategies for a given job title.

```python
print (chain_tools.invoke({'job title' : 'Data Scientist'}))
```

```python
print (chain_strategy.invoke({'tools' : '''1. Python
2. R Programming
3. SQL
4. Tableau
5. Hadoop'''}))
```

## Combine Multiple Chains
Use RunnablePassthrough to pass outputs between chains and create a multi-step workflow.

```python
chain_tools = chat_template_tools|chat|string_parser| {"tools":RunnablePassthrough()}
chain_strategy = chat_template_strategy|chat|string_parser
```

```python
combined_chain = chain_tools|chain_strategy
```

```python
print (combined_chain.invoke({"job title": 'Data Scientist'}))
```
